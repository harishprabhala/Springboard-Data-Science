```python
%matplotlib inline
import numpy as np
import pandas as pd
import nltk
import string
import matplotlib.pyplot as plt
import scipy.sparse as sparse
import spacy
import seaborn as sns
nlp = spacy.load('en')
```

```python
pet = pd.read_json('/Users/harishprabhala/Downloads/Pet_Supplies_5.json', lines=True)
```

```python
from gensim import corpora, models
from gensim.corpora import Dictionary, MmCorpus
from gensim.models.ldamulticore import LdaMulticore
from nltk.corpus import stopwords
import re
```

```python
stop = stopwords.words('english')
```

```python
colname = pet.reviewText
```

```python
col = colname.astype(str).apply(lambda x:x.split())
```

```python
col = [w for w in col if not w in stop]
```

```python
dictionary = corpora.Dictionary(col)
```

```python
corpus = [dictionary.doc2bow(text) for text in col]
```

```python
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
lda = LdaMulticore(corpus_tfidf, id2word=dictionary, num_topics=20, workers=3)
```

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis=pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```
